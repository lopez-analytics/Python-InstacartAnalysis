### Create a bar chart from the “loyalty_flag” column.¶

```python
bar_2 = df_combo['loyalty_flag'].value_counts().plot.bar(color=['#0AAD05', '#FF7009', '#003D29'], edgecolor='black')

# Set titles 
plt.title('Number of Customer by Loyalty Status')
plt.xlabel('Loyalty Status')
plt.ylabel('Customers')

# Format y-axis tick labels with commas
bar_2.yaxis.set_major_formatter(mtick.StrMethodFormatter('{x:,.0f}'))

# Adjust figure size and use tight layout
plt.gcf().set_size_inches(10, 6)
plt.tight_layout()

# Rotate x-axis labels for better readability
plt.xticks(rotation=45)

plt.show()
```

```python

# Grouping the data by loyalty status and order ID, and calculating the total amount spent per order
loyalty_order_totals = df_combo.groupby(['loyalty_flag', 'order_id'])['prices'].sum().reset_index()

# Calculating the average amount spent per order for each loyalty status
average_spent_per_order = loyalty_order_totals.groupby('loyalty_flag')['prices'].mean()


# Creating a bar chart to visualize the reordered average amount spent per order for each loyalty status
bar_3= average_spent_per_order.plot(kind='bar',color=['#003D29','#FF7009','#0AAD05'], edgecolor='black')
plt.xlabel('Loyalty Status')
plt.ylabel('Average Amount Spent per Order')
plt.title('Average Amount Spent per Order by Loyalty Status')
plt.tight_layout() 
plt.show()
```

### Create a scatterplot to check for a relationship between age and income using the sns.scatterplot() function. In a markdown cell beneath your scatterplot, describe what the scatterplot shows.¶

```python
# Create a scatterplot

# Create a custom color palette
custom_palette = ['#0AAD05', '#FF7009', '#FFFFFF']

# Assuming df_combo is your DataFrame
scatterplot_2 = sns.scatterplot(x='age', y='income', data=df_combo, palette=custom_palette)
# Set title for the plot
plt.title('Ages by Income')
# Set titles for the axes
plt.xlabel('Income')
plt.ylabel('Age')
# Show the plot
plt.show()
```

Across the board, the majority of users reside in the 200,000 and under range. However, there are significant differences after the age of 40. Many more users are in the 200,00 - 300,000 range. Also, only after 40 are there any users with income above 400,000.
