### Find aggregated mean of 'number_of_orders' col grouped by 'department_id' for entire datasetÂ¶


```python
# Groupby on ords_prods_merge 'department_id', aggregate mean # of orders

opm_derived.groupby('department_id').agg({'number_of_orders':['mean']})
```

A few things are different about the output of this aggregation on the entire dataset vs on the subset (in Exercise). Some of the values are different, such as Produce (4). In the subset it has a mean of 18.83, in the whole dataframe the mean is 17.81. We are also only presented the rows that appear on record in the subset (and as they appear, not in chronological order), whereas we are guaranteed all departments if aggregating the entire dataset.

### Create a loyalty flag for existing customers using the transform() and loc() functions

```python
# Groupby on dataset by user_id, apply transform() on the ânumber_of_orders" col to generate max orders per user

opm_derived['max_order'] =  opm_derived.groupby(['user_id'])['number_of_orders'].transform(np.max)
```

```python
opm_derived.head()
```

```python
opm_derived.shape
```

```python
# Use new max_orders col to create loyalty flag

opm_derived.loc[opm_derived['max_order'] > 40, 'loyalty_flag'] =  'Loyal customer'
```

```python
opm_derived.loc[(opm_derived['max_order'] <= 40) & (opm_derived['max_order'] > 10), 'loyalty_flag'] = 'Regular customer'
```

```python
opm_derived.loc[opm_derived['max_order'] <= 10, 'loyalty_flag'] = 'New customer'
```

```python
# Check output

opm_derived.shape
```

```python
opm_derived.head()
```

```python
opm_derived['loyalty_flag'].value_counts(dropna = False)
```

### Determine whether the prices of products purchased by loyal customers differ from those purchased by regular or new customers.

```python
# Aggregate mean, min, max of prices of products by loyalty category

opm_derived.groupby('loyalty_flag').agg({'prices': ['mean', 'min', 'max']})
```

### Create 'type_of_spender' col to target different customers in marketing campaigns

```python
# Create avg_spend col

opm_derived['avg_spend'] = opm_derived.groupby(['user_id'])['prices'].transform(np.mean)
```

```python
# Check output

opm_derived.shape
```

```python
opm_derived.head()
```

```python
# Create spender flag

opm_derived.loc[opm_derived['avg_spend']<10,'spender_flag']='Low_spender'
```

```python
opm_derived.loc[opm_derived['avg_spend']>= 10, 'spender_flag'] = 'High_spender'
```

```python
# Check spender flag output

opm_derived['spender_flag'].value_counts(dropna = False)
```

```python
opm_derived[['user_id', 'avg_spend', 'spender_flag', 'prices']].head(60)
```

### Create an order frequency flag that marks the regularity of a userâs ordering behavior according to the median in the âdays_since_prior_orderâ column.Â¶

```python
# Create a frequent user col, Groupyby user_id, aggreagate days_since_prior_order median

opm_derived['customer_frequency'] = opm_derived.groupby(['user_id'])['days_since_prior_order'].transform(np.median)
```

```python
# Check output

opm_derived.head(50)
```

```python
opm_derived.shape
```

```python
# Create customer frequency flag

opm_derived.loc[opm_derived['customer_frequency'] > 20, 'frequency_flag'] = 'Non-frequent Customer'
```

```python
opm_derived.loc[(opm_derived['customer_frequency']>10) & (opm_derived['customer_frequency']<=20),'frequency_flag'] = 'Regular customer'
```

```python
opm_derived.loc[opm_derived['customer_frequency'] <= 10, 'frequency_flag'] = 'Frequent Customer'
```

```python
# Check flag output
opm_derived['frequency_flag'].value_counts(dropna = False)
```

```python
opm_derived[['user_id', 'customer_frequency', 'frequency_flag']].head(50)
```
