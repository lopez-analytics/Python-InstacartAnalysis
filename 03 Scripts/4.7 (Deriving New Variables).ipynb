## 02. 4.7 Exercise

```python
# Create subset of dateframe (1 million rows)

df = ords_prods_merge[:1000000]
```

```python
# Define price labeling function

def price_label(row):
    
    if row['prices'] <=5:
        return 'Low-range product'
    elif (row['prices'] >5) and (row['prices'] <= 15):
        return 'Mid-range product'
    elif row['prices'] > 15:
        return 'High range product'
    else: return 'Not enough data'
```

```python
# Apply created price labeling function

df['price_range_elif'] = df.apply(price_label, axis = 1)
```

```python
# Product price label counts

df['price_range_elif'].value_counts(dropna = False)
```

```python
# Investigate why no high-range prods (>15) were returned

df['prices'].max()
```

#### Using loc instead of if/else

```python
# Define loc conditions

df.loc[df['prices'] > 15, 'price_range_loc'] = 'High range product'
```

```python
df.loc[(df['prices'] <= 15) & (df['prices'] > 5), 'price_range_loc'] = 'Mid range product'
```

```python
df.loc[df['prices'] <= 5, 'price_range_loc'] = 'Low range product'
```

### Create busiest_days column

```python
result_2 = []

for value in ords_prods_merge["order_day_of_week"]:
    if value == 0 or value == 1:
            result_2.append("Busiest days")
    elif value == 4 or value == 3:
            result_2.append("Slowest days")
    else:
            result_2.append("Regularly busy")
```

### Create busiest_period_of_day column

```python
# Check for busiest/slowest hours 

ords_prods_merge['order_hour_of_day'].value_counts(dropna = False)
```

```python
# Create for-loop to sort top 5, bottom 5, and average

result_3 = []

for value in ords_prods_merge["order_hour_of_day"]:
    if value == 10 or value == 11 or value == 14 or value == 15 or value ==13:
            result_3.append("Most orders")
    elif value == 4 or value == 3 or value == 2 or value ==5 or value == 1:
            result_3.append("Fewest orders")
    else:
            result_3.append("Average orders")
```

```python
ords_prods_merge['busiest_period_of_day'] = result_3
```
